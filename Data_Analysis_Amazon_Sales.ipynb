# *Transformacion de los datos*
Una vez terminada la limpieza, lo que voy a hacer es transformar mi dataset con el objetivo de sacar la informacion necesaria para responder las preguntas que me plantee al inicio


```python
#Lo primero que voy a hacer es renombrar las
# columnas para que tengan un nombre mas amigable
df_final.rename(columns= {"title": "producto", "rating":"calificacion",
"number_of_reviews": "cantidad_de_reseñas",
"bought_in_last_month": "comprados_el_ultimo_mes", "current/discounted_price": "precio_actual",
 }, inplace=True)

df_final = df_final.rename(columns= {"title": "producto", "rating": "calificacion",
"number_of_reviews": "cantidad_de_reseñas",
"bought_in_last_month": "comprados_el_ultimo_mes", "current/discounted_price": "precio_actual",
 })
```

```python
df_final.head(10)
```

```python
#Ahora lo que voy a hacer es transformar los valores a numeros para
#poder manejarlos

# Tomo todo como string, uso split() para separar por espacios y
# luego tomo el primer indice(la calificacion en este caso)
df_final['calificacion'] = df_final['calificacion'].astype(str).str.split().str[0]

# Convierto a número decimal (float)
df_final['calificacion'] = df_final['calificacion'].astype(float)
```

```python
# Elimino las comas de la columna 'Cantidad_de_reseñas'
df_final['cantidad_de_reseñas'] = df_final['cantidad_de_reseñas'].astype(str).str.replace(',', '')

# Convierto a número decimal (float)
df_final['cantidad_de_reseñas'] = df_final['cantidad_de_reseñas'].astype(float)
```

```python
# Elimino las comas de la columna 'precio_actual'
df_final['precio_actual'] = df_final['precio_actual'].astype(str).str.replace(',', '')

# Convierto a número decimal (float)
df_final['precio_actual'] = df_final['precio_actual'].astype(float)
```

```python
#Ahora la parte mas complicada, estandarizar la columna "comprados_el_ultimo_mes"
#Detecte que esta columna tiene algunos valores sin utilidad como
# list, list_price o :, asi que voy a filtrarlos

def dato_util(valor):
    texto = str(valor).lower()
    # Si tiene 'list' o ':', devuelve False (No lo quiero)
    if 'list' in texto or ':' in texto:
        return False
    # Si no tiene eso, devuelve True (Sí lo quiero)
    else:
        return True

# Aplico la función a toda la columna
datos_buenos = df_final['comprados_el_ultimo_mes'].apply(dato_util)

df_final = df_final[datos_buenos]
```

```python
#Como siguen quedando valores basura, filtro solo aquellas filas
#que contengan un digito numerico
df_final = df_final[df_final['comprados_el_ultimo_mes'].astype(str).str.strip().str.match(r'^\d')]

#ahora lo que hago es la transformacion final para quedarme con
#valores numericos

def compras(valor):
    # Convierto a string para manejarlo
    valor = str(valor)

    # Limpio el texto extra " bought in past month " y el "+"
    valor = valor.replace('bought in past month', '').replace("+", '')


    # Lógica de multiplicadores (K = 1000, M = 1.000.000)
    if 'K' in valor:
        # Quito la K y multiplico por 1000
        numero = float(valor.replace('K', ''))
        return numero * 1000
    elif 'M' in valor:
        # Quito la M y multiplico por 1 millón
        numero = float(valor.replace('M', ''))
        return numero * 1000000
    else:
        # Si es solo un número, lo devuelvo sin modificarlo
        return float(valor)

# Aplico la función a toda la columna
df_final['comprados_el_ultimo_mes'] = df_final['comprados_el_ultimo_mes'].apply(compras)
```

```python
#Finalmente, para mejorar la legibilidad de los graficos, voy a
#acortar los nombres de los productos hasta la primera coma.

def acortar(texto):
    # 1. Convierto a string
    texto_str = str(texto)

    # 2. Uso split para dividir el texto donde haya comas
    # Y despues tomo el elemento [0] (la primera parte)
    nombre_corto = texto_str.split(',')[0]

    return nombre_corto

# Aplico la función a la columna 'producto'
df_final['producto'] = df_final['producto'].apply(acortar)
```

# *Visualizacion de los datos*
En esta sección procedo a responder las preguntas que me hice, mediante la utilizacion de graficos y tablas




#  **¿Cuales son los productos que mas ingresos generaron el ultimo mes?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

#  Creo la columna de ingresos
# Ingresos = Precio unitario * Cantidad vendida
df_final['ingresos_mensuales'] = df_final['precio_actual'] * df_final['comprados_el_ultimo_mes']

#  Obtengo los 10 productos con mayores ingresos
top_productos = df_final.nlargest(10, 'ingresos_mensuales')

# Muestro la tabla escrita de resultados
print("--- TOP 10 PRODUCTOS POR INGRESOS ---\n")
print(top_productos[['producto', 'ingresos_mensuales']])

# Visualización (Gráfico de Barras)
plt.figure(figsize=(6, 6))

sns.barplot(
    data=top_productos,
    y='producto',
    x='ingresos_mensuales',
    palette='viridis',
    hue='producto',
    legend=False
)

plt.title('Productos que generaron más ingresos el último mes', fontsize=16)
plt.xlabel('Ingresos Estimados (Millones de dolares)', fontsize=16)
plt.ylabel('')
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.show()
```

**Conclusion**: En el gráfico y en la tabla se puede observar claramente cuales son los productos cuyas ventas mas ingresos generaron el ultimo mes. Los ingresos del top 10 van desde casi 2 millones de dolares hasta los casi 5 millones.


# **¿Y de los 5 con mas ingresos, que porcentaje se llevo cada uno?**

```python

#  Filtro el top 5 (ya tengo el calculo de ingresos mensuales)
top_5_ingresos = df_final.nlargest(5, 'ingresos_mensuales')

#Genero el grafico circular
plt.figure(figsize=(7, 7))

plt.pie(
    top_5_ingresos['ingresos_mensuales'],
    labels=top_5_ingresos['producto'],
    autopct='%1.1f%%', # Muestro el porcentaje con 1 decimal
    startangle=140,    # Roto el gráfico para que se vea mejor
    colors=sns.color_palette('viridis', 5), # Uso la paleta de Seaborn
)
plt.title('Participación de Ingresos: Top 5 Productos ', fontsize=20, fontweight='bold')
plt.axis('equal') # Aseguro que el grafico sea un circulo

plt.show()
```

**Conclusión**: El grafico circular nos demuestra claramente que los porcentajes entre los productos que mas ingresos generaron el último mes en la plataforma estan bastante bien distribuidos, lo cual nos clarifica que no hay una especie de "monopolio" donde un solo producto genera muchisimas mas ganancias que el resto .

# **¿Que productos tienen mejor rating en promedio, los mas caros, los mas baratos o los de un precio medio?**

```python

# Creo categorias de precio
# Divido los datos en 3 partes iguales: Barato (33%), Medio (33%), Caro (33%)

df_final['rango_precio'] = pd.qcut(
    df_final['precio_actual'],
    q=3,
    labels=['Barato', 'Medio', 'Caro']
)

#  Agrupo y promedio
# Agrupo por el nuevo rango y calculo la media de la calificación
analisis_precios = df_final.groupby('rango_precio', observed = True)['calificacion'].mean().reset_index()

# Mouestro los datos calculados
print("--- Promedio de Rating por Categoría ---")
print(analisis_precios)

# Visualización
plt.figure(figsize=(8, 5))

# Gráfico de barras para comparar categorías
sns.barplot(
    data=analisis_precios,
    x='rango_precio',
    y='calificacion',
    palette='viridis',
    hue='rango_precio',
    legend=False
)

plt.title('Rating Promedio según el Rango de Precio', fontsize=14)
plt.xlabel('Categoría de Precio', fontsize=12)
plt.ylabel('Calificación Promedio (1-5)', fontsize=12)

# Ajusto el eje Y entre 3 y 5 para notar mejor las diferencias visuales
plt.ylim(3, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

**Conclusion**: Este grafico nos muestra que los productos que entran en la categoria  "baratos" tienen, en promedio, las mejores calificaciones, seguidos por los productos de precio "medio" y por ultimo los "caros".
Este analisis nos da a entender que la gente en general le da bastante importancia al precio de un producto al momento de elegir cuan satisfecha esta con una compra.
